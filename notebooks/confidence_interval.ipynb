# Confidence Intervals for Accuracy, Precision, Recall, and F-Measure

```python
import numpy as np
from scipy.stats import norm
```

```python
alpha_default = 0.05            # default type I error
n             = range(1, 150+1) # range of sample sizes
numReps       = 50000           # repetitions of experiments
```

Check empirical coverage of confidence interval

```python
def coverage(true_value, sampler, estimator, ci, n, alpha=alpha_default, numReps=numReps):
    res = np.empty(numReps)*np.NaN
    samples = sampler(n, numReps)
    estimates = estimator(samples, n)
    cis = ci(estimates, n)
    res = [lower < true_value < upper for (lower, upper) in cis] 
    return mean(res)
```

## Definition of Confidence Intervals

```python
def naive(rHat, n, alpha=alpha_default):
    #norm.interval(1.0-alpha, loc=X.mean(), scale=X.std()/sqrt(n))
    z_alpha = norm.ppf(1-alpha/2.0)
    stderr  = sqrt(rHat*(1-rHat)/n)
    width   = z_alpha*stderr
    return [(p-z, p+z) for (p, z) in zip(rHat, width)]
```

```python
def wilson(rHat, n, alpha=alpha_default):
    z_alpha     = norm.ppf(1-alpha/2.0)
    z_alpha2    = z_alpha*z_alpha
    stderrTilde = sqrt(rHat - rHat*rHat + z_alpha2 / (4*n)) / (sqrt(n) + z_alpha2 / sqrt(n))
    rTilde      = (rHat + z_alpha2/(2*n)) / (1 + z_alpha2/n)
    width       = z_alpha * stderrTilde
    return [(p-z, p+z) for (p, z) in zip(rTilde, width)]
```

```python
def agresti_coull(rHat, n, alpha=alpha_default):
    z_alpha     = norm.ppf(1-alpha/2.0)
    z_alpha2    = z_alpha*z_alpha
    nTilde      = n + z_alpha2
    rTilde      = (rHat*n + z_alpha2/2.0) / nTilde
    stderrTilde = sqrt(rTilde*(1.0-rTilde) / nTilde)
    width       = z_alpha * stderrTilde
    return [(p-z, p+z) for (p, z) in zip(rTilde, width)]
```

## Coverage of Confidence Intervals for Binomial Proportion (e.g., Accuracy)

```python
acc = 0.05

def binomial_sampler(n, reps):
    return numpy.random.binomial(n, acc, size=reps)
    
def accuracy(samples, n):
    return samples / float(n)

def coverage_acc(ci, i):
    return coverage(acc, binomial_sampler, accuracy, ci, i)
```

```python
y_acc_naive  = [coverage_acc(naive, i) for i in n]
y_acc_wilson = [coverage_acc(wilson, i) for i in n]
y_acc_ac     = [coverage_acc(agresti_coull, i) for i in n]
```

```python
plot(n, y_acc_naive, label='naive', color='blue')
plot(n, y_acc_wilson, label='wilson', color='green')
plot(n, y_acc_ac, label='agresti-coull', color='black')
plot(n, [1-alpha_default for _ in n], label='optimal', color='red')
xlabel('num samples n')
ylabel('coverage')
ylim([0.6, 1.0])
xlim([0, n[-1]+1])
title('Empirical Coverage for Accuracy={:.2}'.format(acc))
legend(loc=4);
```

## Coverage of Confidence Intervals for Ratio of Binomial Proportion (e.g., Recall, Precision)

```python
# return ratio of tp, tn, fp, fn for given precision and recall
def get_sample_from_precision_recall(prec, rec):
    tn = 0.0 # true negatives do not have any influence to measures under study
    tp = prec*rec*(1-tn)/(prec+rec-prec*rec);
    fp = 1-tn-tp/rec;
    fn = 1-tn-tp-fp;
    return (tp, tn, fp, fn)
```

```python
prec = 0.95
rec = 0.8

(tp, tn, fp, fn) = get_sample_from_precision_recall(prec, rec)

def multinomial_sampler(n, reps):
    samples = np.random.multinomial(n, [tp, tn, fp, fn], size=reps)
    arr = np.zeros((reps,), dtype=[('TP','float'),('TN','float'),('FP','float'),('FN','float')])
    arr['TP'], arr['TN'], arr['FP'], arr['FN'] = samples.T[0], samples.T[1], samples.T[2], samples.T[3]
    return arr
    
def precision(samples, n):
    return samples['TP']/(samples['TP']+samples['FP'])

def coverage_prec(ci, i):
    return coverage(prec, multinomial_sampler, precision, ci, i)

def recall(samples, n):
    return samples['TP']/(samples['TP']+samples['FN'])

def coverage_rec(ci, i):
    return coverage(rec, multinomial_sampler, recall, ci, i)
```

## Coverage of Confidence Intervals for Harmonic Mean of Ratio of Binomial Proportion (e.g., F-Measure)

```python
f_meas = 2*prec*rec / (prec + rec)

def f_measure(samples, n):
    precs = precision(samples, n)
    recs = recall(samples, n)
    return 2*precs*recs / (precs + recs)

def coverage_f(ci, i):
    return coverage(f_meas, multinomial_sampler, f_measure, ci, i)
```

```python
y_f_naive  = [coverage_f(naive, i) for i in n]
y_f_wilson = [coverage_f(wilson, i) for i in n]
y_f_ac     = [coverage_f(agresti_coull, i) for i in n]
```

```python
plot(n, y_f_naive, label='naive', color='blue')
plot(n, y_f_wilson, label='wilson', color='green')
plot(n, y_f_ac, label='agresti-coull', color='black')
plot(n, [1-alpha_default for _ in n], label='optimal', color='red')
xlabel('num samples n')
ylabel('coverage')
ylim([0.6, 1.0])
xlim([0, n[-1]+1])
title('Empirical Coverage for F-measure={:.2}'.format(f_meas))
legend(loc=4);
```
