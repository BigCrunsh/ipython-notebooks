# Confidence Intervals for Accuracy, Precision, Recall, and F-Measure

## Confidence Intervals for Binomial Proportions (e.g., Accuracy)

```python
import numpy as np
from scipy.stats import norm
```

```python
alpha_default = 0.05
p_default = 0.05
n_default = 100
numReps = 50000
```

```python
def naive(X, alpha=alpha_default):
    #norm.interval(1.0-alpha, loc=X.mean(), scale=X.std()/sqrt(n))
    n = len(X)
    z_alpha = norm.ppf(1-alpha/2.0)
    rHat = X.mean(axis=0)
    stderr = X.std(axis=0)/sqrt(n)
    width = z_alpha*stderr
    return [(p-z, p+z) for (p, z) in zip(rHat, width)]
```

```python
def wilson(X, alpha=alpha_default):
    n = len(X)
    z_alpha = norm.ppf(1-alpha/2.0)
    z_alpha2 = z_alpha*z_alpha
    rHat = X.mean(axis=0)
    stderrCirc = sqrt(rHat - rHat*rHat + z_alpha2 / (4*n)) / (sqrt(n) + z_alpha2 / sqrt(n))
    rCirc = (rHat + z_alpha2/(2*n)) / (1 + z_alpha2/n)
    width = z_alpha * stderrCirc
    return [(p-z, p+z) for (p, z) in zip(rCirc, width)]
```

```python
def agresti_coull(X, alpha=alpha_default):
    n = len(X)
    z_alpha = norm.ppf(1-alpha/2.0)
    z_alpha2 = z_alpha*z_alpha
    nTilde = n + z_alpha2
    rTilde = (X.sum(axis=0) + z_alpha2/2.0) / nTilde
    stderrTilde = sqrt(rTilde*(1.0-rTilde) / nTilde)
    width = z_alpha * stderrTilde
    return [(p-z, p+z) for (p, z) in zip(rTilde, width)]
```

```python
def coverage(ci, p=p_default, n=n_default, numReps=numReps):
    alpha=0.05
    res = np.empty(numReps)*np.NaN
    X = (rand(n, numReps) < p)
    cis = ci(X)
    res = [lower < p < upper for (lower, upper) in cis] 
    return mean(res)
```

```python
x = range(1, 150+1)
y_naive = [coverage(naive, p=p_default, n=i) for i in x]
y_wilson = [coverage(wilson, p=p_default, n=i) for i in x]
y_ac = [coverage(agresti_coull, p=p_default, n=i) for i in x]
```

```python
plot(x, y_naive, label='naive', color='blue')
plot(x, y_wilson, label='wilson', color='green')
plot(x, y_ac, label='agresti-coull', color='black')
plot(x, [1-alpha_default for _ in x], label='optimal', color='red')
xlabel('num samples n')
ylabel('coverage')
ylim([0.6, 1.0])
xlim([0, x[-1]+1])
legend(loc=4);
```
