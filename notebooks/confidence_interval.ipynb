```python
alpha_default = 0.05            # default type I error
n             = range(1, 150+1) # range of sample sizes
num_reps      = 50000           # repetitions of experiments
```

```python
def coverage(true_value, sampler, ci, estimator, num_draws, alpha=alpha_default, num_reps=num_reps):
    res = np.empty(num_reps)*np.NaN
    samples = sampler(num_draws, num_reps)
    cis = ci(estimator(samples))
    res = [lower < true_value < upper for (lower, upper) in cis] 
    return mean(res)
```

## Coverage of Confidence Intervals for Binomial Proportion (e.g., Accuracy)

```python
# ground truth accuracy
acc = 0.99

# samples confusion matrices for given accuracy
# tn and fn are set to zero (segmentation doesn't matter)
def acc_sampler(n, reps):
    samples = np.random.multinomial(n, [acc, 0.0, 1.0-acc, 0.0], size=reps)
    arr = np.zeros((reps,), dtype=[('TP','float'),('TN','float'),('FP','float'),('FN','float')])
    arr['TP'], arr['TN'], arr['FP'], arr['FN'] = samples.T[0], samples.T[1], samples.T[2], samples.T[3]
    return arr

# shortcut for accuracy estimates
def coverage_acc(ci_type, i):
    return coverage(acc, acc_sampler, ci_type, ci.AccuracyEstimator, i)
```

Accuracy: Check empirical coverage of confidence intervals

```python
y_acc_wald  = [coverage_acc(ci.NormalCIs.ci_wald, i) for i in n]
y_acc_wilson = [coverage_acc(ci.BinomialPropertionCIs.ci_wilson, i) for i in n]
y_acc_ac     = [coverage_acc(ci.BinomialPropertionCIs.ci_agresti_coull, i) for i in n]
```

```python
plot(n, y_acc_wald, label='wald', color='blue')
plot(n, y_acc_wilson, label='wilson', color='green')
plot(n, y_acc_ac, label='agresti-coull', color='black')
plot(n, [1-alpha_default for _ in n], label='optimal', color='red')
xlabel('num samples n')
ylabel('coverage')
ylim([0.6, 1.0])
xlim([0, n[-1]+1])
title('Empirical Coverage for Accuracy={:.2}'.format(acc))
legend(loc=4);
```

## Coverage of Confidence Intervals for Ratio of Binomial Proportion (e.g., Recall, Precision)

```python
# ground truth precision / recall
prec = 0.95
rec = 0.8
eta = 0.3
fmeas = (prec*rec) / (eta*rec+(1-eta)*prec)
tn = 0.0 # true negatives do not have any influence to measures under study

# return ratio of tp, tn, fp, fn for given precision and recall
def get_sample_from_precision_recall(prec, rec):
    tp = prec*rec*(1-tn)/(prec+rec-prec*rec);
    fp = 1-tn-tp/rec;
    fn = 1-tn-tp-fp;
    return (tp, tn, fp, fn)

# samples confusion matrices for given prec/rec
# tn is set to zero (segmentation doesn't matter)
def prec_rec_sampler(n, reps):    
    (tp, tn, fp, fn) = get_sample_from_precision_recall(prec, rec)
    samples = np.random.multinomial(n, [tp, tn, fp, fn], size=reps)
    arr = np.zeros((reps,), dtype=[('TP','float'),('TN','float'),('FP','float'),('FN','float')])
    arr['TP'], arr['TN'], arr['FP'], arr['FN'] = samples.T[0], samples.T[1], samples.T[2], samples.T[3]
    return arr

# shortcut for precision estimates
def coverage_prec(ci_type, i):
    return coverage(prec, prec_rec_sampler, ci_type, ci.PrecisionEstimator, i)

# shortcut for recall estimates
def coverage_rec(ci_type, i):
    return coverage(rec, prec_rec_sampler, ci_type, ci.RecallEstimator, i)

# shortcut for f-measure estimates
def coverage_f(ci_type, i):
    def f_estimator(sample):
        return ci.FMeasureEstimator(sample, eta)
    return coverage(fmeas, prec_rec_sampler, ci_type, f_estimator, i)
```
